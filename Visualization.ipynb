```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
tips=pd.read_csv("datasets/tips.csv")
tips
```

### Do people mostly come to the restaurant for dinner?

```python
fig, ax = plt.subplots()

ax.hist(tips.time, bins=3)

ax.set_xlabel("time")
ax.set_ylabel("count")
plt.show()
```

### Is there a correlation between the number of people at the table and the total bill?

```python
corr_tips=tips.corr()

sns.heatmap(corr_tips)
plt.show()
```

### Which variables have the most impact on the tip left? Let's visualize the data.

```python
sns.relplot(x="tip", y="total_bill", data=tips)
plt.show()
```

```python
sns.relplot(x="tip", y="total_bill", data=tips, hue="size")
plt.show()
```

```python
sns.relplot(x="tip", y="total_bill", data=tips, hue="size", kind="scatter", style="sex")
plt.show()
```

```python
sns.relplot(x="tip", y="total_bill", data=tips, hue="size", kind="scatter", style="sex", size="day", sizes=(15, 200), legend="full")
plt.show()
```

### Relationship between tip and size

```python
sns.scatterplot(x="size", y="tip", data=tips, hue="sex")
plt.show()
```

### Relationship between tip and total_bill

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="size")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips)
plt.show()
```
