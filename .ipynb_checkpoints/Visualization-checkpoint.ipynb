```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
tips=pd.read_csv("tips.csv")
tips
```

```python
fig, ax = plt.subplots()
ax.hist(tips.time, bins=4)
ax.set_xlabel("time")
ax.set_ylabel("count")
plt.show()
```

```python
corr_tips=tips.corr()
sns.heatmap(corr_tips)
plt.show()
```
